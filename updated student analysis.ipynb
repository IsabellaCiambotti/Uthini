```python
stud_summary = pd.read_csv('Copy of Newer Analysis - Student Summary.csv')
```

```python
pd.options.display.max_columns = 1000
stud_summary.head()
```

```python
# filter out students that aren't required to use Uthini
stud_summary = stud_summary[stud_summary['Result'] != 'Exempt']
```

```python
# sort by Channel ID for proper indexing when merging
stud_summary.sort_values(by=['Channel ID'], ascending = True, inplace = True)
```

```python
stud_summary = stud_summary.reset_index(drop=True);
```

```python
# merge relevant columns from the two dataframes on Channel ID
stud_full = pd.merge(stud, stud_summary[['Student Number','Telegram ID','First Name','Last Name','GEMP','Mobile number',
                                         'Email','Result','Channel ID']], on=['Channel ID'])
```

```python
stud_full.head()
```

```python
# create a new column that sorts weeks remaining by GEMP membership
# use df to call full dataframe because it's shorter and faster ;)
weeks = []
df = stud_full

for row in df['GEMP']: 
    if row ==1.0:
        weeks.append(11)
    elif row==2.0:
        weeks.append(9)
    else:
        weeks.append('error')
        
df['Weeks Remaining'] = weeks
```

```python
# this statistic works better for visualization when rounded
df['avg sess/week'] = df['avg sess/week'].round(2)
```

```python
df.head()
```

### perform column-wise calcuations:
- projected sessions at current session rate
- projected chunks at current chunk and session rate
- how many chunks need to be completed in the remaining weeks at current chunk and session rate
- how many sessions need to be completed in the remaining weeks at current chunk and session rate

```python
df['Projected sessions'] = df['avg sess/week'] * df['Weeks Remaining']
```

```python
df['Projected chunks'] = df['avg chunks/sess'] *df['Projected sessions']
```

```python
#df['Session deficit'] = df['remaining sess'] - df['Projected sessions'] This isn't used here but could be useful. 
df['Chunk deficit'] = df['remaining chunks'] - df['Projected chunks']
```

```python
df['Required chunks per week'] = df['Chunk deficit']/df['Weeks Remaining']
```

```python
df['Required sessions at chunk rate']= df['Chunk deficit']/df['avg chunks/sess']
```

```python
df['Required sessions per week']=df['Required sessions at chunk rate']/df['Weeks Remaining']
```

```python
df.head()
```

```python
# flag students who will go over the session limit if they stay at their projected session and chunk rate
over_sessions = []
over_sessions = (df['Required sessions per week']* df['Weeks Remaining']) > df['remaining sess']
over_sessions.value_counts()
```

```python
# add that as a column
df['Over session limit?'] = over_sessions
```

### edit some of the columns to make reporting clearer to students:
- if students have a negative 'required sessions per week' or 'required chunks per week' measurement, they're on track and don't need to show up as flagged in a report/visualization. 
- create a new column, "session rate," that will let students know approx. how many sessions they do a week
- create a progress percentage column for an easier understanding of chunk progress

```python
req1 = []

for row in df['Required sessions per week']: 
    if row <0:
        req1.append(0)
    else: 
        req1.append(row)
req1;
req1 = [round(x) for x in req1]
```

```python
df['Required sessions per week'] = req1
```

```python
df['Session rate'] = df['Required sessions per week'] - df['avg sess/week'].astype(int)
```

```python
req4 = []
for row in df['Session rate']: 
    if row > 0:
        req4.append(row)
    elif row < 0:
        req4.append('0')
    else:
        req4.append('0')
req4;
```

```python
df['Session rate'] = req4
```

```python
req5 = []
for row in df['Required chunks per week']: 
    if row > 0:
        req5.append(row)
    elif row < 0:
        req5.append('0')
    else:
        req5.append('0')
req5;

req6 = []
for row in df['Required sessions at chunk rate']: 
    if row > 0:
        req6.append(row)
    elif row < 0:
        req6.append('0')
    else:
        req6.append('0')
req6;
```

```python
df['Required sessions at chunk rate'] = req6
df['Required chunks per week'] = req6
```

```python
df["Progress Percentage"] = (df['current chunk']/650 *100).round(2)
```

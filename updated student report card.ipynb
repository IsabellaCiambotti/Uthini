# generating a spreadsheet geared for student clarity

just paring down the larger analysis datasheet so it's useable as a "student report card"

```python
import pandas as pd
import numpy as np
```

```python
report = pd.read_csv('updated_student_analysis.csv')
```

```python
pd.options.display.max_columns = 1000
report.head()
```

drop unneccessary columns, reformat decimals, and rename columns to be clearer

```python
report.rename(columns={'Session rate':'Increase sessions by'}, inplace=True);
```

```python
report = report[['Last Name','First Name','GEMP','Result','Weeks Remaining','Progress Percentage'
                 ,'Increase sessions by','Over session limit?']]
```

```python
report['Progress Percentage'] = report['Progress Percentage'].round(0).astype(int)
```

```python
report.head()
```

```python
report.to_csv('student_report_card.csv')
```
